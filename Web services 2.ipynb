```python
import json 
import requests
import csv
from os import listdir
import openpyxl
```

```python
response  = requests.get('https://www.metaweather.com/api/location/1521894/').text
json_response = json.loads(response)
data = json_response['consolidated_weather']
```

```python
row = [data[0]['applicable_date'],data[0]['the_temp'],data[0]['humidity'],data[0]['visibility'],data[0]['air_pressure'],data[0]['wind_speed']]
headers = ['datetime', 'weather temperature', 'humidity', 'visibility', 'air pressure', 'wind speed']
excel_names = ['A1','B1','C1','D1','E1','F1']
```

```python
def getweather_data_inCsv():
    if 'weather_data.csv' not  in listdir('.'):
        with open('weather_data.csv','w',newline='') as f:
            writer = csv.writer(f)
            writer.writerow(['datetime', 'weather temperature', 'humidity', 'visibility', 'air pressure', 'wind speed'])
            writer.writerow(row)
    else:
        with open('weather_data.csv','a',newline='') as f:
            writer = csv.writer(f)
            writer.writerow(row)
                
```

```python
getweather_data_inCsv()
```

```python
#note : i got the three days from the data i fetched throw json once but its not right i know but i cant use pandas for now and cant append in 
def write_toXlsx():
    if 'weather_data.xlsx' not  in listdir('.'):
        xlsx = openpyxl.Workbook()
        sheet = xlsx.active
        for i in range(len(headers)):
            sheet[excel_names[i]] = headers[i]
        sheet.append(row)
        xlsx.save('weather_data.xlsx')
    else:
        xlsx = openpyxl.load_workbook('weather_data.xlsx')

        ## getting the sheet to active
        sheet = xlsx.active
        sheet.append(row)
        xlsx.save('weather_data.xlsx')
```

```python
write_toXlsx()
```
