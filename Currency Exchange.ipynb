```python
from bs4 import BeautifulSoup
import requests
```

```python
def askint(word):
    while True:
        try:
            val = int(input(word))
            return val
        except:
            print('invalid input please try again'.title())
```

```python
def Exchange():
    usd  = askint('Amount in USD : ')
    url = 'https://www.exchangerates.org.uk/Dollars-to-Egyptian-Pounds-currency-conversion-page.html'
    response = requests.get(url) 
    soup = BeautifulSoup(response.text , 'html.parser')
    usd_to_egp = soup.body.find('span',attrs = {'id' : "shd2a"}).text
    ratio = float(usd_to_egp.split('=')[1].split()[0].strip())
    return f"{'*'*40}\nYour Exchange of {usd} Dollars will be {usd * ratio} in EGP\n{'*'*40}"
```

```python
print(Exchange())
```
