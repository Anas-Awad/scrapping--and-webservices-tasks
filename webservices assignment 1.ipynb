```python
import requests
import csv
import json
```

```python
def askint(word):
    while True:
        try:
            val = int(input(word))
            return val
        except:
            print('invalid input please try again'.title())

    
```

```python
def exchange():
    usd = askint('Amount of dollars want to exchange: ')
    response = json.loads(requests.get(f'https://api.frankfurter.app/latest?amount={usd}&from=USD&to=EUR').text)
    return '{}\ndate : {} \nAmount in Dollars {} = {} in EUROS\n{}'.format('*'*40,response['date'],response['amount'],response['rates']['EUR'],'*'*40)
```

```python
print(exchange())
```
